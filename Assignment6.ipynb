```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv ")
```

```python
data = data.dropna()
```

```python
sns.distplot(data['Rainfall'], bins= 20)
```

```python
sns.kdeplot(data['Sunshine'], shade=True)
```

```python
sns.distplot(data['WindGustSpeed'], hist=False)
```

```python
sns.countplot(data['RainToday'])
```

```python
sns.boxplot(data['MaxTemp'])
```

```python
series = pd.Series(data['MinTemp'])
sns.distplot(series)
```

```python
sns.distplot(series, kde=False, rug=True)
```

```python
series = pd.Series(data['MaxTemp'])
sns.kdeplot(series, shade=True)
```

```python
sns.distplot(series, hist=False, rug=True)
```

```python
data = pd.DataFrame(data, columns=['Rainfall', 'Evaporation'])
sns.jointplot(x = 'Rainfall', y = 'Evaporation', data = data)
```

```python
series = pd.Series(data['Evaporation'])
sns.distplot(series)
```

```python
data = pd.DataFrame(data, columns=['Rainfall', 'Evaporation'])
sns.jointplot(x='Rainfall', y='Evaporation', data=data, kind='kde')
```

```python
data= pd.DataFrame(data, columns=['Rainfall', 'Evaporation'])
with sns.axes_style('white'):
    sns.jointplot(x='Rainfall', y='Evaporation', data=data, kind='hex')
```

```python
tip = sns.load_dataset('tips')
sns.pairplot(tip)
```

```python
sns.pairplot(tip, hue='time')
```

```python
sns.stripplot(x = 'day', y='total_bill', data=tip)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tip)
```
